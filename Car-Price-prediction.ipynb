```python
def change_price_to_numeric(r):
    return float(("".join(r.split(" ")[0])).replace(",",""))
```

```python
data2['Average price'] = data2['Average price'].astype(str).apply(change_price_to_numeric)
data2['Minimum price'] = data2['Minimum price'].astype(str).apply(change_price_to_numeric)
data2['Maximum price'] = data2['Maximum price'].astype(str).apply(change_price_to_numeric)
```

```python
data.drop_duplicates(inplace=True)

data.reset_index(drop=True)
```

```python
data["Car Brand"] = data['Car Model'].apply(lambda x:"".join(x.split()[0]))
data["Car_Model_Year"] = data['Car Model'].apply(lambda x:"".join(x.split()[-1]))
data[['Car Brand',"Car_Model_Year"]]
```

```python
data['Old Price']= data['Old Price'].apply(lambda r:r[:-4].replace(',',"")).astype("float")

data['New Price']= data['New Price'].apply(lambda r:r[:-4].replace(',',"")).astype("float")

data['Price Change'] = data['New Price'] - data['Old Price']
data['Price Change']
```

```python
data['date_range']=pd.to_datetime(data['date_range'],format="%d/%m/%Y",errors="coerce")

data['Month'] = data['date_range'].dt.month_name()
data['Month_Num'] = data['date_range'].dt.month
data['Quarter'] = data['date_range'].dt.quarter.astype(str).apply(lambda r:r.split(".")[0])
data['Year'] = data['date_range'].dt.year.astype(str).apply(lambda r:r.split(".")[0])
```
